```python
import torch
import matplotlib.pyplot as plt
import os
import sys
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(project_root)

from models.MRF import ising_model

# 示例 n x n 矩阵（取值为 -1 或 1）
n = 100
J = 0.44
iter = 9
ising_model = ising_model.Ising(n, J)
matrices = ising_model.stable_Gibbs_sampling(iter)

fig, axes = plt.subplots(2, 5, figsize=(15, 3))  # 宽度大一些

# 遍历并可视化每个矩阵
for i, ax in enumerate(axes.flatten()):
    if i < iter+1:
        img_data = (matrices[i] + 1) // 2
        ax.imshow(img_data, cmap="bwr", interpolation="nearest")
        ax.set_title(f"Matrix {i+1}")
    ax.axis("off")

# 添加共享 colorbar（可选）
# fig.colorbar(im, ax=axes.ravel().tolist(), shrink=0.6, label="Value")

plt.tight_layout()
plt.show()
```
